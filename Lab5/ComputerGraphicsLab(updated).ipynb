# 2D Translation

**Problem Statement:**

Given a square with coordinates:  
A(0, 3), B(3, 3), C(3, 0) and D(0, 0).  

Apply a translation of:
- **1 unit along the x-axis**
- **1 unit along the y-axis**

Find the **new coordinates** of the square after translation.


```python
import matplotlib.pyplot as plt

# Original square coordinates
square = [(0, 3), (3, 3), (3, 0), (0, 0)]
labels_orig = ['A', 'B', 'C', 'D']

# Translation values
tx, ty = 1, 1

# Apply translation
translated = [(x + tx, y + ty) for x, y in square]
labels_trans = ["A'", "B'", "C'", "D'"]

# Close the shapes
square_closed = square + [square[0]]
translated_closed = translated + [translated[0]]

# Plot original square
plt.plot(*zip(*square_closed), label='Original Square', color='blue')

# Label original points with coordinates
for (x, y), label in zip(square, labels_orig):
    plt.text(x, y, f" {label}({x},{y})", fontsize=10, color='blue', fontweight='bold')

# Plot translated square
plt.plot(*zip(*translated_closed), '--', label='Translated Square', color='green')

# Label translated points with coordinates
for (x, y), label in zip(translated, labels_trans):
    plt.text(x, y, f" {label}({x},{y})", fontsize=10, color='green', fontweight='bold')

# Labels and settings
plt.title('Square Translation with Point Labels and Coordinates')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

# 2D Scaling

**Problem Statement:**

Given a square with the following coordinates:  
- A(0, 3)  
- B(3, 3)  
- C(3, 0)  
- D(0, 0)  

Apply a **scaling transformation** with:  
- Scale factor **2** along the x-axis  
- Scale factor **3** along the y-axis  

Find the **new coordinates** of the object after scaling.


```python
import matplotlib.pyplot as plt

# Original square coordinates
square = [(0, 3), (3, 3), (3, 0), (0, 0)]
labels_orig = ['A', 'B', 'C', 'D']

# Scaling parameters
sx, sy = 2, 3

# Apply scaling
scaled = [(x * sx, y * sy) for x, y in square]
labels_scaled = ["A'", "B'", "C'", "D'"]

# Close shapes
square_closed = square + [square[0]]
scaled_closed = scaled + [scaled[0]]

# Plot original square
plt.plot(*zip(*square_closed), label='Original Square', color='blue')
for (x, y), label in zip(square, labels_orig):
    plt.text(x, y, f" {label}({x},{y})", fontsize=10, color='blue', fontweight='bold')

# Plot scaled square
plt.plot(*zip(*scaled_closed), '--', label='Scaled Square', color='red')
for (x, y), label in zip(scaled, labels_scaled):
    plt.text(x, y, f" {label}({x},{y})", fontsize=10, color='red', fontweight='bold')

# Settings
plt.title('Square Scaling (Sx=2, Sy=3)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

 # 2D Rotation

**Problem Statement:**

Given a triangle with coordinates:
- A(0, 0)
- B(1, 0)
- C(1, 1)

Rotate the triangle **90 anticlock-wise about the origin (0, 0)**.

Find the **new coordinates** after rotation.


```python
import matplotlib.pyplot as plt
import math

# Original triangle coordinates
triangle = [(0, 0), (1, 0), (1, 1)]
labels_orig = ['A', 'B', 'C']

# Rotation angle
angle = 90
theta = math.radians(angle)

# Apply rotation (anticlockwise around origin)
rotated = [
    (round(x * math.cos(theta) - y * math.sin(theta), 2),
     round(x * math.sin(theta) + y * math.cos(theta), 2))
    for x, y in triangle
]
labels_rot = ["A'", "B'", "C'"]

# Close shapes
triangle_closed = triangle + [triangle[0]]
rotated_closed = rotated + [rotated[0]]

# Plot original triangle
plt.plot(*zip(*triangle_closed), label='Original Triangle', color='blue')
for (x, y), label in zip(triangle, labels_orig):
    plt.text(x, y, f" {label}({x},{y})", fontsize=10, color='blue', fontweight='bold')

# Plot rotated triangle
plt.plot(*zip(*rotated_closed), '--', label=f'Rotated {angle}', color='red')
for (x, y), label in zip(rotated, labels_rot):
    plt.text(x, y, f" {label}({x},{y})", fontsize=10, color='red', fontweight='bold')

# Settings
plt.title('Triangle Rotation (90 Anticlockwise)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()

# Print new coordinates
print("New Coordinates after Rotation:")
for lbl, coord in zip(labels_rot, rotated):
    print(f"{lbl}: {coord}")
```

# 2D Reflection

**Problem Statement**

Given a triangle with vertices:

A = (3,4)

B = (6,4)

C = (5,6)

Apply reflection over:

1.X-axis

2.Y-axis

```python
import matplotlib.pyplot as plt

# Original triangle coordinates
triangle = [(3, 4), (6, 4), (5, 6)]
labels_orig = ['A', 'B', 'C']

# Reflection over X-axis: (x, y) -> (x, -y)
reflected_x = [(x, -y) for x, y in triangle]
labels_rx = ["A_x", "B_x", "C_x"]

# Reflection over Y-axis: (x, y) -> (-x, y)
reflected_y = [(-x, y) for x, y in triangle]
labels_ry = ["A_y", "B_y", "C_y"]

# Close the triangles
triangle_closed = triangle + [triangle[0]]
reflected_x_closed = reflected_x + [reflected_x[0]]
reflected_y_closed = reflected_y + [reflected_y[0]]

# Plot original triangle
plt.plot(*zip(*triangle_closed), label='Original Triangle', color='blue')
for (x, y), label in zip(triangle, labels_orig):
    plt.text(x, y, f"{label}({x},{y})", color='blue', fontsize=10, fontweight='bold')

# Plot reflection over X-axis
plt.plot(*zip(*reflected_x_closed), '--', label='Reflected over X-axis', color='green')
for (x, y), label in zip(reflected_x, labels_rx):
    plt.text(x, y, f"{label}({x},{y})", color='green', fontsize=10, fontweight='bold')

# Plot reflection over Y-axis
plt.plot(*zip(*reflected_y_closed), '--', label='Reflected over Y-axis', color='red')
for (x, y), label in zip(reflected_y, labels_ry):
    plt.text(x, y, f"{label}({x},{y})", color='red', fontsize=10, fontweight='bold')

# Settings
plt.title('Triangle Reflection')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()

# Print new coordinates
print("Reflection over X-axis:")
for lbl, coord in zip(labels_rx, reflected_x):
    print(f"{lbl}: {coord}")

print("\nReflection over Y-axis:")
for lbl, coord in zip(labels_ry, reflected_y):
    print(f"{lbl}: {coord}")
```

# 2D shearing

Problem Statement

Given a triangle with vertices:
- **A** = (1,1)  
- **B** = (0,0)  
- **C** = (1,0)  

Apply a **shear transformation** with:
- Shear factor **2** along the x-axis  
- Shear factor **2** along the y-axis  

```python
import matplotlib.pyplot as plt

# Original triangle coordinates
triangle = [(1, 1), (0, 0), (1, 0)]
labels_orig = ['A', 'B', 'C']

# Shear parameters
shx, shy = 2, 2

# Apply shear along X-axis
sheared_x = [(x + shx * y, y) for x, y in triangle]
labels_sx = ["A_x", "B_x", "C_x"]

# Apply shear along Y-axis
sheared_y = [(x, y + shy * x) for x, y in triangle]
labels_sy = ["A_y", "B_y", "C_y"]

# Close the triangles for plotting
triangle_closed = triangle + [triangle[0]]
sheared_x_closed = sheared_x + [sheared_x[0]]
sheared_y_closed = sheared_y + [sheared_y[0]]

# Plot original triangle
plt.plot(*zip(*triangle_closed), label='Original Triangle', color='blue')
for (x, y), label in zip(triangle, labels_orig):
    plt.text(x, y, f"{label}({x},{y})", color='blue', fontsize=10, fontweight='bold')

# Plot shear along X-axis
plt.plot(*zip(*sheared_x_closed), '--', label='Sheared along X-axis', color='green')
for (x, y), label in zip(sheared_x, labels_sx):
    plt.text(x, y, f"{label}({x},{y})", color='green', fontsize=10, fontweight='bold')

# Plot shear along Y-axis
plt.plot(*zip(*sheared_y_closed), '--', label='Sheared along Y-axis', color='red')
for (x, y), label in zip(sheared_y, labels_sy):
    plt.text(x, y, f"{label}({x},{y})", color='red', fontsize=10, fontweight='bold')

# Settings
plt.title('Triangle Shearing')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()

# Print new coordinates
print("Sheared along X-axis:")
for lbl, coord in zip(labels_sx, sheared_x):
    print(f"{lbl}: {coord}")

print("\nSheared along Y-axis:")
for lbl, coord in zip(labels_sy, sheared_y):
    print(f"{lbl}: {coord}")
```
