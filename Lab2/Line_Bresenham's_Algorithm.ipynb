```python
import matplotlib.pyplot as plt
```

```python
def draw_line_bba(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    x, y = x1, y1
    points = [(x, y)]

    p = 2 * dy - dx

    for _ in range(dx):
        if p < 0:
            p = p + 2 * dy
        else:
            y += 1
            p = p + 2 * dy - 2 * dx
        x += 1
        points.append((x, y))

    return points
```

```python

def plot_line(points, title):
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, marker='o')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

```

```python
# Example usage:
x1, y1 = 9, 18
x2, y2 = 14, 22
```

```python
bba_points = draw_line_bba(x1, y1, x2, y2)
plot_line(bba_points, 'Bresenham Line Drawing')
```
